Encoding

```python
import pandas as pd

s = pd.Series(list('abca'))
print(f"before dummies", s)
pd.get_dummies(s)
print(f"after dummies", pd.get_dummies(s))
```

Feature scaling

```python
# Estandarization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
print(f"Data before estandarization", data)
scaler.fit(data)
print(f"Data after estandarization", scaler.transform(data))

# Normalization 
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()
data = [[4, 1, 2], [1, 3, 9], [0, 1, 5]]
print(f"Data before normalization", data)
normalizer.fit(data)
print(f"Data after normalization", normalizer.transform(data))
```
