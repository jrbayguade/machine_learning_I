# Pre-procesamiento de datos y análisis descriptivo básico

**Importación de librerías**

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Librarías cargadas correctamente.")
```

**Carga de datos**

```python
# Cargar el dataset desde sklearn
data = datasets.load_breast_cancer()

# Claves del dataset
print(f"Keys in the dataset: {list(data.keys())}")

# Descripción general del dataset
print(data['DESCR'][:3500])
```

**Explorar atributos del dataset**

```python
# Nombres de los atributos (features)
print('Atributos del dataset:')

i = 1
for feature in data.feature_names:
    print(f"{i}. {feature}")
    i += 1
```

```python
# Clases objetivo (target)
print(f"Clases objetivo (target): {data.target_names}")

print(f"0 = {data.target_names[0]}")
print(f"1 = {data.target_names[1]}")
```

**Guardar atributos y respuesta en variables X e Y**

Por convención, usamos **X** (mayúscula) para la matriz de atributos e **y** (minúscula) para el vector objetivo.

```python
# Crear las variables X e Y
X = data.data
y = data.target

print(f'Tipo de X: {type(X)}')
print(f'Tipo de y: {type(y)}')
```

**Contamos filas y columnas**

```python
# Dimensiones del dataset
print(f'Número de muestras (filas): {X.shape[0]}')
print(f'Número de atributos (columnas): {X.shape[1]}')
print(f'Dimensiones de X: {X.shape}')
print(f'Dimensiones de y: {y.shape}')
```

```python
# Distribución de las clases

# Creamos un DataFrame con 
df_clases = pd.DataFrame({
    'clase': data.target_names,
    'codigo': [0, 1],
    'muestras': np.bincount(y),
    'porcentaje': np.bincount(y) / len(y) * 100
})

print(df_clases.to_string(index=False))
```

**Convertir a DataFrame de Pandas**

Trabajar con un DataFrame nos facilita la exploración y manipulación de los datos.

```python
# Crear un DataFrame con los datos
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

# Mostrar las primeras filas
df.head()
```

**Estadística descriptiva básica**

```python
df.describe().round(2)
```

**Buscar datos faltantes**

```python
# Verificar valores nulos
nulos_por_columna = df.isnull().sum()
total_nulos = nulos_por_columna.sum()

print(f'Total de valores nulos en el dataset: {total_nulos}')

if total_nulos == 0:
    print('\nEl dataset no tiene valores faltantes.')
else:
    print('\nColumnas con valores nulos:')
    print(nulos_por_columna[nulos_por_columna > 0])
```

```python
# Información general del DataFrame
df.info()
```

# Visualización exploratoria

```python
# Distribución de las clases
fig, ax = plt.subplots(figsize=(6, 4))
colores = ['#E74C3C', '#2ECC71']
conteos = np.bincount(y)

ax.bar(data.target_names, conteos, color=colores)
ax.set_ylabel('Número de muestras')
ax.set_title('Distribución de clases en Breast Cancer Wisconsin')

for i, c in enumerate(conteos):
    ax.text(i, c + 5, str(c), ha='center', fontweight='bold')
    
plt.tight_layout()
plt.show()
```

```python
# Histograma de un atributo: 'mean radius'
fig, ax = plt.subplots(figsize=(8, 4))

for clase, nombre, color in zip([0, 1], data.target_names, colores):
    ax.hist(X[y == clase, 0], bins=20, alpha=0.6, label=nombre, color=color)

ax.set_xlabel('mean radius')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución de mean radius por clase')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Matriz de correlación de los primeros 10 atributos
fig, ax = plt.subplots(figsize=(10, 8))
corr = df.iloc[:, :10].corr()
im = ax.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)

ax.set_xticks(range(10))
ax.set_yticks(range(10))
ax.set_xticklabels(data.feature_names[:10], rotation=45, ha='right', fontsize=8)
ax.set_yticklabels(data.feature_names[:10], fontsize=8)
ax.set_title('Matriz de correlación (primeros 10 atributos)')

plt.colorbar(im, ax=ax, shrink=0.8)
plt.tight_layout()
plt.show()
```
